```python
Con questo esercizio si vuole stimare il valore di pigreco, con relativa incertezza, simulando l'esperimento di Buffon. Fulcro di questo esercizio è il metodo
del data blocking: si noti dal grafico l'evoluzione della stima dopo 100 blocchi, e come il valore calcolato di pigreco converge al valore atteso.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def read_data_from_file(filename):
    # Inizializza due liste vuote per memorizzare i dati
    pi = []
    err = []
    
    # Apre il file e legge i dati riga per riga
    with open(filename, 'r') as file:  # Modalità "r" per la lettura
         # Ignora la prima riga (riga di intestazione)
        next(file)
        for line in file:
            # Divide ogni riga in due numeri e li aggiunge alle rispettive liste
            values = line.split()
            pi.append(float(values[0]))
            err.append(float(values[1]))
            
    return pi, err

# Codice per generare il grafico dopo aver letto i dati dal file
pi, err = read_data_from_file('pi.dat')# Definizione degli indici (x)
x = np.arange(len(pi), dtype=float)
x *= (100/len(pi))
plt.errorbar(x, pi, yerr=err)

# Aggiungi una retta orizzontale con valore costante 3.14, valore atteso
plt.axhline(y=3.14, color='red', linestyle='-', label='Valore atteso di pi')

# Creazione del grafico con valori di pi e barre d'errore
plt.title('Valore di pi')
plt.xlabel('Numero di blocchi')
plt.ylabel('Valore di pi')
plt.grid(True)
plt.legend()
plt.show()
```
