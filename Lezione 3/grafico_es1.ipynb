```python
In questo esercizio è stata svolta una simulazione che mostra che l'evoluzione temporale del prezzo di un'attività sia stocastica
e mostri un moto browniano geometrico (GBM). La simulazione è stata eseguita in due modi differenti: il primo che valuta direttamente il prezzo finale 
dell'attività S(T); il secondo valutando il prezzo dell'attività su intervalli di tempo discreti, ovvero calcolando S(t0), S(t0+1).. fino a S(t100=T=1).
Qui sotto sono presenti dei confronti grafici tra i prezzi calcolati per le opzioni europee di Call e Put in entrambi i metodi, e il valore analitico atteso. 
```

```python
import numpy as np
import matplotlib.pyplot as plt

def read_data_from_file(filename):
    # Inizializza due liste vuote per memorizzare i dati
    mediacalldir = []
    errcalldir = []
    mediaputdir = []
    errputdir = []
    mediacalldisc = []
    errcalldisc = []
    mediaputdisc = []
    errputdisc = []
  

    # Apre il file e legge i dati riga per riga
    with open(filename, 'r') as file:  # Modalità "r" per la lettura
    # Ignora la prima riga (riga di intestazione)
        next(file)
        for line in file:
            # Divide ogni riga in due numeri e li aggiunge alle rispettive liste
            values = line.split()
            mediacalldir.append(float(values[0]))
            errcalldir.append(float(values[1]))
            mediaputdir.append(float(values[2]))
            errputdir.append(float(values[3]))
            mediacalldisc.append(float(values[4]))
            errcalldisc.append(float(values[5]))
            mediaputdisc.append(float(values[6]))
            errputdisc.append(float(values[7]))
            
           
            
    return  mediacalldir, errcalldir, mediaputdir, errputdir, mediacalldisc, errcalldisc, mediaputdisc, errputdisc

# Chiamata alla funzione per leggere i dati dal file
mediacalldir, errcalldir, mediaputdir, errputdir, mediacalldisc, errcalldisc, mediaputdisc, errputdisc = read_data_from_file("callput.dat") 

#Grafico della Call Diretta
x = np.arange(100, dtype=float)  # Specifica il dtype come float
plt.errorbar(x, np.array(mediacalldir), yerr=np.array(errcalldir), color= 'blue', label='Call diretta')
plt.xlabel('Numero di blocchi')
plt.ylabel('Call')
plt.ylim(12.5, 20.5)
plt.axhline(y=14.975790778311286, color='yellow', linestyle='-', label='Analytic Call Solution') 
plt.grid(True)
plt.legend()
plt.show()

#Grafico della Call Discreta
x = np.arange(100, dtype=float)  # Specifica il dtype come float
plt.errorbar(x, np.array(mediacalldisc), yerr=np.array(errcalldisc), color = 'red', label = 'Call discreto')
plt.xlabel('Numero di blocchi')
plt.ylabel('Call')
plt.ylim(13, 17)
plt.axhline(y=14.975790778311286, color='yellow', linestyle='-', label='Analytic Call Solution') 
plt.grid(True)
plt.legend()
plt.show()

#Grafico della Put Diretta
z = np.arange(100, dtype=float)  # Specifica il dtype come float
plt.errorbar(z, np.array(mediaputdir), yerr=np.array(errputdir), color= 'blue', label='Put diretta')
plt.xlabel('Numero di blocchi')
plt.ylabel('Put')
plt.ylim(3.5, 6.5)
plt.axhline(y=5.4595325819072364, color='yellow', linestyle='-', label='Analytic Put Solution') 
plt.grid(True)
plt.legend()
plt.show()

#Grafico della Put Discreta
z = np.arange(100, dtype=float)  # Specifica il dtype come float
plt.errorbar(z, np.array(mediaputdisc), yerr=np.array(errputdisc), color = 'red', label = 'Put discreto')
plt.xlabel('Numero di blocchi')
plt.ylabel('Put')
plt.ylim(4.5, 6.3)
plt.axhline(y=5.4595325819072364, color='yellow', linestyle='-', label='Analytic Put Solution') 
plt.grid(True)
plt.legend()
plt.show()


#Confronto delle due Call
x = np.arange(100, dtype=float)  # Specifica il dtype come float
plt.errorbar(x, np.array(mediacalldir), yerr=np.array(errcalldir), color= 'blue', label='Call diretta')
plt.errorbar(x, np.array(mediacalldisc), yerr=np.array(errcalldisc), color = 'red', label = 'Call discreto')
plt.xlabel('Numero di blocchi')
plt.ylabel('Call')
plt.ylim(12.5, 20.5)
plt.axhline(y=14.975790778311286, color='yellow', linestyle='-', label='Analytic Call Solution') 
plt.grid(True)
plt.legend()
plt.show()

#Confronto delle due Put
z = np.arange(100, dtype=float)  # Specifica il dtype come float
plt.errorbar(z, np.array(mediaputdir), yerr=np.array(errputdir), color= 'blue', label='Put diretta')
plt.errorbar(z, np.array(mediaputdisc), yerr=np.array(errputdisc), color = 'red', label = 'Put discreto')
plt.xlabel('Numero di blocchi')
plt.ylabel('Put')
plt.ylim(3.5, 6.5)
plt.axhline(y=5.4595325819072364, color='yellow', linestyle='-', label='Analytic Put Solution') 
plt.grid(True)
plt.legend()
plt.show()
```
