```python
In questo esercizio è stato calcolato il valore di un integrale tramite due metodi: il metodo della media, che sfrutta il campionamento di una
distribuzione uniforme, e l'importance sampling, che sfrutta una funzione che approssimi bene l'integranda sull'intervallo di integrazione. 
Si noti dal confronto grafico come è nettamente preferibile il secondo metodo. L'ultimo grafico mostra come la densità di probabilità utilizzata 
nel metodo dell'importance sampling sia una scelta ragionata ed efficace.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def read_data_from_file(filename):
    # Inizializza due liste vuote per memorizzare i dati
    medie = []
    err = []
    mediaHoM = []
    errHoM = []

    # Apre il file e legge i dati riga per riga
    with open(filename, 'r') as file:  # Modalità "r" per la lettura
    # Ignora la prima riga (riga di intestazione)
        next(file)
        for line in file:
            # Divide ogni riga in due numeri e li aggiunge alle rispettive liste
            values = line.split()
            medie.append(float(values[0]))
            err.append(float(values[1]))
            mediaHoM.append(float(values[2]))
            errHoM.append(float(values[3]))
            
    return medie, err, mediaHoM, errHoM

# Chiamata alla funzione per leggere i dati dal file
medie, err, mediaHoM, errHoM = read_data_from_file("ris.dat") 

#Grafica il calcolo dell'integrale col metodo della media
x = np.arange(100, dtype=float)  # Specifica il dtype come float
plt.errorbar(x, np.array(medie), yerr=np.array(err), color = 'blue', label = 'Integrale con metodo Media')
plt.xlabel('Numero di blocchi')
plt.ylabel('Valore integrale')
plt.ylim(0.89, 1.11)
plt.axhline(y=1, color='yellow', linestyle='-', label='Soluzione analitica')
plt.grid(True)
plt.legend()
plt.show()

#Grafica il calcolo dell'integrale col metodo dell'Hit or Miss
x = np.arange(100, dtype=float)  # Specifica il dtype come float
plt.errorbar(x, np.array(mediaHoM), yerr=np.array(errHoM), color= 'red', label='Integrale con importance sampling')
plt.xlabel('Numero di blocchi')
plt.ylabel('Valore integrale')
plt.ylim(0.994, 1.003)
plt.axhline(y=1, color='yellow', linestyle='-', label='Soluzione analitica')
plt.grid(True)
plt.legend()
plt.show()

#Confronta il calcolo dell'integrale con i due metodi
x = np.arange(100, dtype=float)  # Specifica il dtype come float
plt.errorbar(x, np.array(mediaHoM), yerr=np.array(errHoM), color= 'red', label='Integrale con importance sampling')
plt.errorbar(x, np.array(medie), yerr=np.array(err), color = 'blue', label = 'Integrale con metodo Media')
plt.xlabel('Numero di blocchi')
plt.ylabel('Valore integrale')
plt.ylim(0.89, 1.11)
plt.axhline(y=1, color='yellow', linestyle='-', label='Soluzione analitica') 
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definire la funzione pi/2 * cos(pi * x / 2)
def func1(x):
    return (np.pi / 2) * np.cos((np.pi * x) / 2)

# Definire la funzione 3/2 * (1 - x^2)
def func2(x):
    return (np.pi / 2) * (1 - x**2)

# Creare l'array x da 0 a 1
x = np.linspace(0, 1, 500)

# Calcolare i valori delle funzioni
y1 = func1(x)
y2 = func2(x)

# Creare il plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r'$\frac{\pi}{2} \cos\left(\frac{\pi x}{2}\right)$', color='blue')
plt.plot(x, y2, label=r'$\frac{\pi}{2} (1 - x^2)$', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Plot integranda f(x) = $\frac{\pi}{2} \cos\left(\frac{\pi x}{2}\right)$ e  densità di probabilità p(x) = $\frac{\pi}{2} (1 - x^2)$')
plt.legend()
plt.grid(True)
plt.show()
```
